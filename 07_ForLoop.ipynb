```python
s = ["Mahmoud", "And", "Ahmed"]

# using for loop with list of strings
for i in s:
    print(i)
```

```python
s = "Mahmoud"
for i in s:
    print(i)
```

```python
for i in range(0, 10, 2):
    print(i)
```

```python
# Prints all letters except 'e' and 's'

for i in 'MahmoudKamalEgypt':

    if i == 'e' or i == 's':
        continue
    print(i)
```

```python
for i in 'Mahmoud':

    # break the loop as soon it sees 'e'
    # or 's'
    if i == 'e' or i == 's':
        break

print(i)
```

```python
# An empty loop
for i in 'Mahmoud':
    pass
print(i)
```

```python
for i in range(1, 4):
    print(i)
else:  # Executed because no break in for
    print("No Break\n")
```

```python
li = ["eat", "sleep", "repeat"]

for i, j in enumerate(li):
    print (i, j)
```

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```
