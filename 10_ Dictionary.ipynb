```python
data = { "name": "Jake", "age": 22 }
print(data)
```

## - Creating a Dictionary

```python
a = {1: 'Mahmoud', 2: 'Kamal', 3: 'Mohamed'}
print(a)

b = dict(a = "Mahmoud", b = "Ahmed", c = "Good")
print(b)
```

## - Accessing Dictionary Items

```python
d = { "name": "Kat", 1: "Python", (1, 2): [1,2,4] }

# Access using key
print(d["name"])

# Access using get()
print(d.get("name"))
```

## - Adding and Updating Dictionary Items


```python
d = {1: 'Mahmoud', 2: 'Kamal', 3: 'Mohamed'}

# Adding a new key-value pair
d["age"] = 22

# Updating an existing value
d[1] = "Python dict"
print(d)
```

## - Removing Dictionary Items

```python
d = {1: 'Mahmoud', 2: 'Kamal', 3: 'Mohamed', 'age':24}

# Using del 
del d["age"]
print(d)

# Using pop() 
val = d.pop(1)
print(val)

# Using popitem()
key, val = d.popitem()
print(f"Key: {key}, Value: {val}")

# Using clear()
d.clear()
print(d)
```

## - Iterating Through a Dictionary

```python
d = {1: 'Mahmoud', 2: 'Kmaal', 'age':24}

# Iterate over keys
for key in d:
    print(key)

# Iterate over values
for value in d.values():
    print(value)

# Iterate over key-value pairs
for key, value in d.items():
    print(f"{key}: {value}")
```

## - Nested Dictionaries

```python
d = {1: 'Mahmoud', 2: 'Kamal', 3: {'A': 'Welcome', 'B': 'To', 'C': 'ClassRoom'}}
print(d)
```
