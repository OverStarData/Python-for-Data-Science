### Create String

```python
s1 = 'Mahmoud'  
s2 = "Kamal"  
print(s1)
print(s2)
```

### Multi-line Strings

```python
s = """I am Learning
Python String"""
print(s)

s = '''I'm a 
Strong'''
print(s)
```

### Accessing characters in String

```python
s = "Mahmoud"
print(s[0])   
print(s[4])
```

```python
s = "Mahmoud Kamal"
print(s[-10])  
print(s[-5])
```

### String Slicing


```python
s = "Mahmoud"
print(s[1:4])    
print(s[:3])     
print(s[3:])    
print(s[::-1])
```

### String Iteration


```python
s = "Python"
for char in s:
    print(char)
```

### String Immutability


```python
s = "Mahmoud"
s = "M" + s[1:]  
print(s)
```

### Deleting a String

```python
s = "Mahmoud"
del s
```

### Updating a String


```python
s = "hello Mahmoud"
s1 = "H" + s[1:]                  
s2 = s.replace("hello", "hi")  
print(s1)
print(s2)
```

### Common String Methods


```python
s = "Mahmoud"
print(len(s))
```

```python
s = "Hello World"
print(s.upper())
print(s.lower())
```

```python
s = "   Mahmoud   "
print(s.strip())    

s = "Python is fun"
print(s.replace("fun", "awesome"))
```

### Concatenating and Repeating Strings


```python
s1 = "Hello"
s2 = "World"
print(s1 + " " + s2)
```

```python
s = "Hello "
print(s * 3)
```

### Formatting Strings



```python
name = "Jake"
age = 22
print(f"Name: {name}, Age: {age}")
```

```python
s = "My name is {} and I am {} years old.".format( "Mahmoud", 24)
print(s)
```

```python
s = "Mahmoud Kamal"
print("Kamal" in s)
print("Ahmed" in s)
```
