## 1. Numeric Data Types

1. Integers
2. Float
3. Complex

```python
a = 5
print(type(a))

b = 5.0
print(type(b))

c = 2 + 4j
print(type(c))
```

## 2. Sequence Data Types

1. String
2. List Data Type
3. Tuple Data Type



```python
s = 'Welcome to the Python World'
print(s)

# check data type 
print(type(s))

# access string with index
print(s[1])
print(s[2])
print(s[-1]) # -1 refers to the last character, -2 is second last, and so on
```

```python
# Empty list
a = []

# list with int values
a = [1, 2, 3]
print(a)

# list with mixed values int and String
b = ["Ahmed", "For", "Kamal", 4, 5]
print(b)
```

```python
a = ["Ahmed", "For", "Kamal"]
print("Accessing element from the list")
print(a[0])
print(a[2])

print("Accessing element using negative indexing")
print(a[-1])
print(a[-3])
```

```python
# initiate empty tuple
tup1 = ()

tup2 = ('Ahmed', 'Good')
print("\nTuple with the use of String: ", tup2)
```

```python
tup1 = (1, 2, 3, 4, 5)

# access tuple items
print(tup1[0])
print(tup1[-1])
print(tup1[-3])
```

## 4. Set Data Type


```python
# initializing empty set
s1 = set()

s1 = set("MahmoudKamal")
print("Set with the use of String: ", s1)

s2 = set(["Ahmed", "And", "Mahmoud"])
print("Set with the use of List: ", s2)
```

#### -Access Set Items


```python
set1 = set(["Ahmed","And","Mahmoud"]) #Duplicates are removed automatically
print(set1) 

# loop through set
for i in set1:
   print(i, end=" ") #prints elements one by one
  
# check if item exist in set   
print("Ahmed" in set1)
```

## 5. Dictionary Data Type


```python
# initialize empty dictionary
d = {}

d = {1: 'Mahmed', 2: 'And', 3: 'Ahmed'}
print(d)

# creating dictionary using dict() constructor
d1 = dict({1: 'Mahmed', 2: 'And', 3: 'Ahmed'})
print(d1)
```

```python
d = {1: 'Ahmed', 'name': 'And', 3: 'Mohamed'}

# Accessing an element using key
print(d['name'])

# Accessing a element using get
print(d.get(3))
```
